```python
import pandas as pd


dic_search = {}     # 키 = 중분류명 밸류 = 소분류명
dic_main = {}       # 키 = 중분류명 밸류 = 소분류명
datafile = pd.read_csv('data1.csv',index_col = 0, encoding='cp949')
city_code= datafile[datafile['시도코드'] == 11]
all_line_Realize = city_code[['상호명','상권업종대분류명','상권업종중분류명','상권업종소분류명','법정동명','지번주소','층정보']]

def address():      #주소동명을 얻어오는 함수
    while True:
        dong = input("현재 위치 동명을 말씀해주세요.ex) 대치동     ")
        dong_name = all_line_Realize[all_line_Realize['법정동명'] == dong]
        show = dong_name['법정동명'].value_counts()
        if show.nunique() == 0:
            print("서울시에 있는 동이 아닙니다. 다시입력해 주십시오.")
        else:
            print("===========================================")
            print()
            return dong_name

def classify():    #각각의 분류명들을 얻어온 함수
    large_category_name = all_line_Realize[all_line_Realize['상권업종대분류명'] == '음식']
    middle_category_nun = large_category_name['상권업종중분류명'].nunique()
    middle_category_un = large_category_name['상권업종중분류명'].unique()
    group_name_nun = large_category_name['상권업종소분류명'].nunique()
    group_name_un = large_category_name['상권업종소분류명'].unique()
    return  middle_category_nun, middle_category_un,group_name_nun,group_name_un

def search_category():      #분류명들을 dic에 담아주는 함수
    middle_count,middle_array,small_count,small_array = classify()
                            #m_count = 중분류카운트 m_array = 중분류배열
                            #s_count = 소분류카운트 s_array = 소분류배열
    for i in range(0,middle_count):
        m_array_equal = all_line_Realize[all_line_Realize['상권업종중분류명'] == middle_array[i]]
        s_count_count = m_array_equal['상권업종소분류명'].nunique()
        s_count_array = m_array_equal['상권업종소분류명'].unique()
        dic_search[middle_array[i]] = s_count_array
        
    return dic_search


def search_middle(dic):    #중분류명을 찾는함수
    i = 0;
    
    print("원하시는 카테고리를 정해주세요.")
    print("===========================")
    for key,value in dic.items():
        i = i + 1
        print(i, " : " , key , end = "    ")
        if i % 3 == 0:
            print()
    while True:
        m_category_input = input("")
        if m_category_input.isdigit() == True:
            if int(m_category_input) > i:
                print("다시입력해주세요.")
            else:
                return m_category_input
        else:
            print("다시입력해주세요.")
        
    
def search_small(m_input,dic):    #소분류명을 찾는함수
    i = 0;
    
    key = dic.keys()
    key_list = list(key)
    print(key_list[int(m_input)-1],"을(는) 선택하셨습니다.")
    print("===========================")    
    print(key_list[int(m_input)-1],"에 대한 카테고리를 선택해주세요.")
    for key,value in dic.items():
        if key == key_list[int(m_input)-1]:
            value_list = list(value)
            for i in range(0,len(value_list)):
                print(i+1," : " , value_list[i], end = "     ")
                if i % 3 == 0:
                    print()
    print("처음으로이동 = 'R'")
    
    while True:
        s_category_input = input("")
        if s_category_input.isdigit() == True:
            if int(s_category_input) > len(value_list):
                print("다시입력해주세요.")
            else:
                print("=====",value_list[int(s_category_input)-1],"선택하셨습니다.========")
                return s_category_input
        else:
            if s_category_input.upper() == 'R':
                main()
                break
            else:
                print("다시입력해주세요.")
          
        
def list_view(find_address,m_input,s_input,dic):    #뷰어부분을 맡는 함수 
    key = dic.keys()
    key_list = list(key)
    addr_gift_two = find_address[find_address['상권업종중분류명'] == key_list[int(m_input)-1]]
    
    for key,value in dic.items():
        if key == key_list[int(m_input)-1]:
            value_list = list(value)
            
    view = addr_gift_two[addr_gift_two['상권업종소분류명'] == value_list[int(s_input)-1]]
    return view
    
    
    
def main():                                     #메인
    find_address = address()
    dic_main = search_category()
    m_input = search_middle(dic_main)
    s_input = search_small(m_input,dic_main)
    view_Previous = list_view(find_address,m_input,s_input,dic_main)
    view = view_Previous[['상호명','법정동명','지번주소','층정보']]
    if view_Previous['상호명'].nunique() == 0:
        print("데이터가 존재하지 않습니다.")
    else:
        print()
        print()
        print(view.to_string(index=False))

 #==================시작 부분=========================

main()
```
